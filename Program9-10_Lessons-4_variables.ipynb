Важное замечание !!!!
Модуль Turtle умеет работать только с целыми значениями углов и шагов черепашки. Однако, если мы захотим нарисовать семиугольник, то нам нужно будет рассчитать угол так 360/ 7 = 51.4285... то есть угол получается дробный. 
В Питоне есть специальные знаки действий, например опрерация деления без остатка,эта операция не округляет, а просто вычисляет целую часть и отбрасывает дробную 
360/ 7 = 51.4285..
360//7 = 51
Поэтому в этих заданиях мы будем использовать на обычное деление, а деление без остатка
//  - это два знака деления подряд!
