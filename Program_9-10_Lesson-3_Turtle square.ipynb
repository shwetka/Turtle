Наверное тебе надоело, что черепашка оставляет за собой только церную линию. Сейчас я покажу тебе как она умеет менять цвет.

Цвет нужно писать на английском 


```python

t=Turtle()
t.color="red"
t.forward (30)
t.color="blue"
t.forward (40)
```

Наша черепашка умеет двигаться так, чтобы не оставлять за собой след. Для этого ей надо дать команду 
penup() 
В скобках ничего писать не надо!
когда нужно, чтобы черепаха снова  оставляла след, дайте ей команду 
pendown()
Внизу пример, как нарисовать прерывистую линию 

```python

t=Turtle()
t.color="red"
t.forward (30)
t.penup()
t.forward(30)
t.pendown()
t.color="blue"
t.forward (40)
```
