Сколько раз и какие команды повторяются? Попробуем написать цикл.


```python
from mobilechelonian import Turtle
t=Turtle()
t.speed(5)
for i in range (2):
    t.forward(100)
    t.right(45)
    t.forward(100)
    t.right(135)
      
```

Наша программа рисует один ромб, но наша фигура состоит из 8 ромбов. Давайте поробуем нарисовать два. 
На какой угол должна повернуть черепашка? Найдите строчку, в которой черепашка поворачивает после первого ромба и попробуйте побобрать нужный угол. 

```python
from mobilechelonian import Turtle
t=Turtle()
t.speed(5)
for i in range (2):
    t.forward(100)
    t.right(45)
    t.forward(100)
    t.right(135)
     
t.right(45)
for j in range (2):
    t.forward(100)
    t.right(45)
    t.forward(100)
    t.right(135)
         
```

 Когда угол подобран и два ромба нарисованы правильно, пришла пора объединить рисование двух ромбов циклом. У нас плучится влоденный друг в друга цикл.
Будьте внимательны с отступами! Отступы нужны после каждого цикла. 

```python
t=Turtle()
t.speed(5)
for i in range (8):
    for i in range (2):
        t.forward(100)
        t.right(45)
        t.forward(100)
        t.right(135)
    t.right(45)
```

Теперь нарисуем второй и подберем угол. 


```python
from mobilechelonian import Turtle
t=Turtle()
t.speed(5)
for i in range (3):
    t.forward (100)
    t.right (120)
t.right(40)    
for i in range (3):
    t.forward (100)
    t.right (120)    
```

Мы нарисовали два треугольника и пришла пора объединить их рисование циклом. Обратите внимание, что при использовании вложенных циклов нужно использовать в каждом цикле свою переменную. 

```python
from mobilechelonian import Turtle
t=Turtle()
t.speed(5)
for j in range (9):
    for i in range (3):
        t.forward (100)
        t.right (120)
    t.right(40)
```
